```python
def download_file(url, dst_path):
    with urllib.request.urlopen(url) as web_file:
        data = web_file.read()
        with open(dst_path, mode='wb') as local_file:
            local_file.write(data)

b = 'http://163.49.30.82'
dst_path = '/work/a01/hnara/lake_volume/other/test/test.dat'
download_file(b+data_location, dst_path)
```
